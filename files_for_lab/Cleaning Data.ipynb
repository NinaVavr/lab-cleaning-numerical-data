```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Get Data

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
customer_df.info
customer_df.head()
```

# Standardize Column Names

```python
cols = []
for i in range(len(customer_df.columns)):
    cols.append(customer_df.columns[i].lower().replace(' ', '_'))
customer_df.columns = cols

customer_df.head()
```

# Datetime Column

```python
customer_df['effective_to_date']=pd.to_datetime(customer_df['effective_to_date'])
customer_df.dtypes
customer_df
```

# Separating Numerical - Continuous/ Discrete

```python
numerical_df = customer_df.select_dtypes(np.number)
numerical_df
```

```python
continuous_df = numerical_df.select_dtypes(float)
continuous_df
```

```python
discrete_df = numerical_df.select_dtypes(int)
discrete_df
```

# Check for Multi-collinearity

```python
correlations = numerical_df.corr()
correlations
```

```python
fig, ax=plt.subplots(figsize=(10, 8))
ax=sns.heatmap(numerical_df.corr(), annot=True)
plt.show()
```

# Distributions

```python
fig, ax = plt.subplots()
for column in discrete_df.columns:
    plt.gca().set(xlabel= column,ylabel='total_claim_amount')
    sns.histplot(discrete_df[column], bins = 10,kde=True)
    plt.show()
```

```python
fig, ax = plt.subplots()
for column in continuous_df.columns:
    plt.gca().set(xlabel= column,ylabel='Frequency')
    sns.boxplot(continuous_df[column], linewidth= 1.0)
    #sns.histplot(continuous_df[column], bins = 100,kde=True)
    plt.show()
```

```python
#From the plots we conclude that the last points may be the real outliers as the points after the wishkers form a normal line
```

# Dealing With The Outliers 

```python
#Proceed with log as it helps us analyze the data for scaling 
```

```python
log_cont = np.log(continuous_df)
log_cont
```

```python
np.log(customer_df['customer_lifetime_value']).hist()
plt.show()
```

```python
# Checking how many values will be removed if the outliers are removed
iqr = np.nanpercentile(customer_df['customer_lifetime_value'],75) - np.nanpercentile(customer_df['customer_lifetime_value'],25)
upper_limit = np.nanpercentile(customer_df['customer_lifetime_value'],75) + 1.5*iqr
print(upper_limit)

extraordinary_points = customer_df[customer_df['customer_lifetime_value'] >200]
len(extraordinary_points)  # This checks the number of points that will be removed
```

```python
#We observe that the outliers consist of a big part of our data and we won't drop them
```
