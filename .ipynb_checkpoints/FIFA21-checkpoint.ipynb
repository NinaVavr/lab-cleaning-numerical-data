Made a decistion to remove some unnesessary columns that have no influence on a predicted OVA. The sources for the decision making are as follows: https://sofifa.com/calculator?hl=en-US

```python
fifa1 = fifa.drop(columns=['Value', 'Growth', 'Wage', 'ID', 'Name', 'Nationality', 'Club', 'Team & Contract', "Joined", "Loan Date End", "Release Clause", "Contract", "A/W", "D/W"], axis=1)
```

```python
cols = []
for column in fifa1.columns:
    cols.append(column.lower())
cols
fifa1
```

```python
fifa1.columns = cols
print(fifa1.shape)
```

```python
fifa1.select_dtypes(include=np.number)
```

```python
numerical_heads = fifa1.select_dtypes(include=['number']).columns.tolist()
print(numerical_heads)
```

```python
fifa1.select_dtypes(include=object)
```

```python
fifa1['w/f'] = fifa1['w/f'].str.rstrip("★").astype(float)
fifa1['sm'] = fifa1['sm'].str.rstrip("★").astype(float)
fifa1['ir'] = fifa1['ir'].str.rstrip("★").astype(float)
fifa1
```

```python
# fifa1['ls'] = fifa1['ls'].apply(lambda x:x.split("+")[0])
```

```python
cols = ['ls','st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']
for col in cols:
    fifa1[col] = fifa1[col].apply(lambda x:x.split("+")[0])
```

```python
fifa1[cols] = fifa1[cols].apply(pd.to_numeric, errors='coerce')
```

```python
# 'def get_first_two(string):
#     return int(string[:2])
# fifa1['ls'] = fifa1['ls'].apply(get_first_two)'
```

```python
fifa1
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(fifa1.dtypes)
```

```python
fifa1['height'] = (fifa1['height'].str.extract(r'(\d+)\'\s*(\d+)"').astype(float).mul([12*2.54, 2.54]).sum(axis=1))
```

```python
fifa1['weight'] = fifa1['weight'].str.rstrip("lbs").astype(float)/2.2
```
